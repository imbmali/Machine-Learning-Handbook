## Handling Imbalanced Dataset

1. Up Sampling
2. Down Sampling

```python
# Imports two popular Python libraries

# NumPy (np) → used for numerical operations and random number generation

# Pandas (pd) → used for working with data in tabular form (DataFrames)

import numpy as np
import pandas as pd

# Sets a fixed random seed

# This ensures that every time you run the code, random values generated by NumPy will be the same

# Very important for experiments, ML models, and debugging

# Set the random seed for reproducibility
np.random.seed(123)

# Create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0,n_class_1
```

```python
## CREATE MY DATAFRAME WITH IMBALANCED DATASET

# This is a classic way to create an imbalanced dataset for ML demos.


# Generates values for feature_1

# np.random.normal() creates data from a normal (Gaussian) distribution

# loc=0 → mean = 0

# scale=1 → standard deviation = 1

# size=n_class_0 → number of samples (e.g., 900)

# So, feature_1 values are centered around 0


# Creates the target/label column

# [0] * n_class_0 means:

# Repeat 0 exactly n_class_0 times

# All rows in this DataFrame belong to Class 0

class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})



class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})


# You created two clusters of data

# Class 0

# Many samples

# Features centered around (0, 0)

# Class 1

# Fewer samples

# Features centered around (2, 2)

# Result: an imbalanced dataset that is:

# Easy to visualize

# Perfect for teaching classification, imbalance problems, SMOTE, ROC-AUC, precision/recall
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)

# pd.concat([class_0, class_1])

# Combines two DataFrames vertically (row-wise)

# class_0 rows come first, followed by class_1

# Resulting DataFrame will have:

# Columns: feature_1, feature_2, target

# Rows: n_class_0 + n_class_1 (e.g., 1000 rows)

# At this stage, the index would look like:

# Class 0 → 0 to 899

# Class 1 → 0 to 99
# (duplicate index values)


# reset_index(drop=True)

# Resets the index to a clean continuous range

# New index becomes 0 to 999

# drop=True means:

# Do not keep the old index as a column

# If you didn’t use drop=True, Pandas would add an extra column called index.

# Final Result

# df is now a single, clean DataFrame

# Shape: (1000, 3)

# Columns:

# feature_1

# feature_2

# target

# Dataset is:

# Imbalanced

# Ready for EDA, visualization, or ML modeling
```

```python
# df.tail()
df.head()
```

```python
df['target'].value_counts()
```

```python
## upsampling
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
# Imports the resample function from scikit-learn

# Used for bootstrapping, upsampling, and downsampling

# Very common in handling imbalanced datasets

from sklearn.utils import resample
df_minority_upsampled=resample(df_minority,replace=True, #Sample With replacement
         n_samples=len(df_majority),
         random_state=42
        )
```

```python
df_minority_upsampled.shape
```

```python
df_minority_upsampled.head()
```

```python
df_upsampled=pd.concat([df_majority,df_minority_upsampled])
```

## Down Sampling

```python
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(123)

# Create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0

class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})

df = pd.concat([class_0, class_1]).reset_index(drop=True)

# Check the class distribution
print(df['target'].value_counts())
```

```python
## downsampling
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
from sklearn.utils import resample
df_majority_upsampled=resample(df_minority,replace=True, #Sample With replacement
         n_samples=len(df_majority),
         random_state=42
        )
```
